```python
BASE_PATH = '/home/masa1357/Dockerdata/gitfile/kaggle_clrp_1st_place_solution/'
INPUT_PATH = BASE_PATH + 'CourceFormat/'
```

```python
import pandas as pd
import os

# Create an empty list to store the dataframes
dfs = []

# Loop through the range of files and read them into the list
for n in range(1, 16):
    file_name = f"cource{n}.csv"
    dfs.append(pd.read_csv(os.path.join(INPUT_PATH,file_name)))

# Concatenate all the dataframes in the list into a single dataframe
df = pd.concat(dfs, ignore_index=True)

df
```

```python
df.columns
```

```python
# Drop the 'userid' column and then concatenate the remaining columns with <b> token
df['excerpt'] = df.drop(columns=['id']).apply(lambda row: '\n'.join(row.dropna().astype(str)), axis=1)

# Display the resulting dataframe with the new 'combined_text' column
df = df[['id', 'excerpt','target']]
df.head()
```

```python
df.to_csv(os.path.join(BASE_PATH,'connected.csv'))
```
