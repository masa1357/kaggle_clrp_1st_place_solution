# User_idごとにデータを結合する

```python
import pandas as pd
import os
```

```python
BASE_PATH = '/home/masa1357/Dockerdata/gitfile/kaggle_clrp_1st_place_solution'
#cn = 3 #course number
```

```python
df = pd.read_csv(os.path.join(BASE_PATH,'translated_df.csv'))
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

df
```

```python
dfs = {}
final_dfs = []
for i in range(1, 16):  # 1から15まで
    dfs[f'df_{i}'] = df[df['course_number'] == i].copy()

    # pivot操作を使用してデータを一行に統合
    pivot_df = dfs[f'df_{i}'].pivot(index='userid', columns='question_number', values='translated_content').reset_index()

    # target列を追加
    target_df = dfs[f'df_{i}'][['userid', 'target']].drop_duplicates().set_index('userid')
    pivot_df = pivot_df.join(target_df, on='userid')
    
    final_dfs.append(pivot_df)
```

```python
final_dfs[14]
```

```python
# 元のデータセットのtranslated_contentの合計数
original_count = df['translated_content'].count()

# 各変換後のデータフレームのtranslated_contentの合計数
transformed_count = sum([dfm.iloc[:, 1:-1].count().sum() for dfm in final_dfs])

original_count, transformed_count, original_count == transformed_count
```

```python
for i, dfm in enumerate(final_dfs, 1):
    dfm.to_csv(f'cource{i}.csv', index=False)
```
