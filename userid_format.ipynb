# User_idごとにデータを結合する

```python
import pandas as pd
import os
```

```python
BASE_PATH = '/home/masa1357/Dockerdata/gitfile/kaggle_clrp_1st_place_solution'
#cn = 3 #course number
```

```python
df = pd.read_csv(os.path.join(BASE_PATH,'translated_df.csv'))
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
unique_courses = df['course_number'].unique()

dfs = {}
for course in unique_courses:
    dfs[f'df_{course}'] = df[df['course_number'] == course]
    
dfs_names_and_heads = {name: df_head for name, df_head in dfs.items()}
unique_courses
```

```python
arr = len(unique_courses)
for i in range(1,arr+1):
    df = dfs[f'df_{i}']
    df['question_col'] = 'translated_content' + df['course_number'].astype(str) + '-' + df['question_number'].astype(str)
    df_pivot = df.pivot(index='userid', columns='question_number', values='translated_content').reset_index()
    df_pivot = pd.merge(df_pivot, df[['userid', 'target']], on='userid', how='left')
    df_pivot.rename(columns={'userid': 'id'}, inplace=True)
    df_pivot.to_csv(f'cource{i}.csv', index=False)
    
```
